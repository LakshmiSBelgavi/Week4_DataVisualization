```python
d4 = pd.read_csv('/content/Metadata_Country.csv')
```

```python
# Converting column names to lowercase for case-insensitive matching
d1.columns = d1.columns.str.lower()
d2.columns = d2.columns.str.lower()
d3.columns = d3.columns.str.lower()
d4.columns = d4.columns.str.lower()

# Merge d1 with d4 based on common columns
d1 = pd.merge(d1, d4[['country name', 'country code', 'region']], on=['country name', 'country code'], how='left')

# Merge d2 with d4 based on common columns
d2 = pd.merge(d2, d4[['country name', 'country code', 'region']], on=['country name', 'country code'], how='left')

# Merge d3 with d4 based on common columns
d3 = pd.merge(d3, d4[['country name', 'country code', 'region']], on=['country name', 'country code'], how='left')

# Print the updated DataFrames
print("Updated d1:")
print(d1.head())

print("\nUpdated d2:")
print(d2.head())

print("\nUpdated d3:")
print(d3.head())
```

```python
#Combining d1 and d2 data frames

# Convert column names to lowercase for case-insensitive matching
d1.columns = d1.columns.str.lower()
d2.columns = d2.columns.str.lower()

# Merge d1 and d2 based on common columns
f_s = pd.merge(d1, d2, on=['country name', 'country code', 'year'], how='inner')

# Print the resulting DataFrame
print(f_s)

```

```python
#Combining first and second dataframe with third one
# Convert column names to lowercase for case-insensitive matching
d3.columns = d3.columns.str.lower()
f_s.columns = f_s.columns.str.lower()

# Merge d3 and f_s based on common columns
f_s_t = pd.merge(d3, f_s, on=['country name', 'country code', 'year'], how='inner')

# Print the resulting DataFrame
print(f_s_t)
```

```python
f_s_t.columns
```

```python
f_s_t = f_s_t.drop('region_y', axis=1)
```

```python
f_s_t = f_s_t.rename(columns={'region_x': 'region'})
```

```python
f_s_t.columns
```

```python


# Drop duplicate columns
f_s_t = f_s_t.loc[:, ~f_s_t.columns.duplicated()]

# Define the desired column order
desired_column_order = ['country name', 'country code', 'region', 'year','population','fertility rate','life expectancy']

# Reorder the columns in f_s_t
f_s_t = f_s_t[desired_column_order]

# Print the resulting DataFrame
print(f_s_t)

```

```python
f_s_t.to_csv('/content/final_dataset.csv', index=False)
```

```python
df = pd.read_csv("/content/final_dataset.csv")
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df.shape
```

```python
import plotly.express as plt
```

```python
# Scatter plot of Fertility Rate VS Life Expectancy rate Region-Wise for every country over the years

plt.scatter(df,
            x='fertility rate',
            y='life expectancy',
            animation_frame="year",
            animation_group='country code',
            size='population',
            hover_name='country code',
            color='region',
            log_x=True,
            size_max=55,
            title='Fertility Rate vs Life Expectancy rate-REGION WISE',
            range_x=[1,10],
            range_y=[10,100])
```

```python
# Population Trends (Region vs Population)

plt.bar(df,
            x='region',
            y='population',
            animation_frame="year",
            animation_group='country code',
            color='region',
            hover_name='country code',
        title='Year Vs Population',
            range_y=[0,2500000000])
```

```python
#Fertility rate distribution
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 4))

ax.hist(df['fertility rate'], bins=20, edgecolor='black',color='purple')

ax.set_xlabel('Fertility Rate')
ax.set_ylabel('Counts (in billions)')
ax.set_title('Fertility Rate Distribution')

plt.show()
```

```python
#Life expectancy Distribution

fig, ax = plt.subplots(figsize=(8, 4))

ax.hist(df['life expectancy'], bins=20, edgecolor='black',color='skyblue')

ax.set_xlabel('life expectancy')
ax.set_ylabel('Counts (in billions)')
ax.set_title('Life Expectancy Distribution')

plt.show()

```

```python
#Correlational analysis
import seaborn as sns
import matplotlib.pyplot as plt

selected_columns = ['life expectancy', 'fertility rate', 'population', 'year']


correlation_matrix = df[selected_columns].corr()

plt.figure(figsize=(10, 4))


heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

plt.title('Correlation Analysis Heatmap')

# Show the plot
plt.show()
```

```python
#Geographical data with fertility rate[Regional Analysis]
import plotly.express as plt
plt.choropleth(df,
                           featureidkey='properties.name',
                           locations='country code',
                           color='fertility rate',
                           hover_name='region',
                           animation_frame='year',
                      animation_group='country code',

               title="Fertility rate [Regional Analysis]"
                          )
```
